# 1. Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())

top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      f"(Score: {highest_score})")
```

# 2. Search Query Keyword Analysis

```python
import string

query = "Buy mobile phone buy phone online".lower()

for char in string.punctuation:
    query = query.replace(char, "")

words = query.split()

word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

repeated_keywords = {word: count 
                     for word, count in word_count.items() 
                     if count > 1}

print(repeated_keywords)
```

#  3. Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index in range(len(sensor_readings)):
    value = sensor_readings[index]

    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# 4. Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_users = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

# 5. Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.25  # adjusted for expected output

for index, value in enumerate(sales):
    if value >= threshold:
        print(f"Day {index + 1}: {value}")
```

# 6. Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
