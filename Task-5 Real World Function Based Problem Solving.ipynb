# 1. Social Media Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like

    # Check status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Print result
    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes = [200, 300, 150, 400]

analyze_engagement(likes)
```

# 3. Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(rainfall_list, required_level):
    total = 0

    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

rainfall_data = [70, 75, 68, 80, 67]
required = 70

check_rainfall(rainfall_data, required)
```

# 4. Social Media – Duplicate Account Detection

```python
def check_duplicates(usernames):
    
    unique_usernames = set(usernames)
    
    if len(unique_usernames) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

accounts = ["rutuja01", "amit23", "rutuja01", "priya99"]

check_duplicates(accounts)
```

# 6. Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(price_list):
    
    premium = []

    for price in price_list:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

crop_prices = [1500, 2500, 1800, 3200, 1900]

filter_premium_crops(crop_prices)
```

# 7. System Monitoring – Application Health

```python
def check_application_health(error_count):

    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)

errors = 7
check_application_health(errors)
```

# 9. E-Learning – Student Attendance Eligibility 

```python
def check_exam_eligibility(attendance_list):

    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance = ["P", "P", "A", "P", "P"]

check_exam_eligibility(attendance)
```

# 10. Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):

    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5

    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

units_consumed = 250

total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
